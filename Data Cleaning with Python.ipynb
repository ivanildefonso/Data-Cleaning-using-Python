# 1.Standardisation

During standardisation, we can rename columns, convert values to lowercase/uppercase. 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('http://bit.ly/DataCleaningDataset', ';')
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_').str.replace('[','').str.replace(']','')
df.head()
```

```python
df2 = pd.read_csv('http://bit.ly/DataCleaningDataset', ';')
df.head()
```

```python
df2.columns = ['name', 'city', 'country', 'height', 'weight', 'account_a', 'account_b','total_account']
df2.head()
```

```python
df['city'] = df['city'].str.lower()
df.head()
```

```python
df['height'] = df['height'] * 2.54
df.head()
```

```python
df.dtypes
```

```python
df['height'] = df['height'].apply(np.int64)
df['account_b'] = df['account_b'].apply(np.float64)
df['total_account'] = df['total_account'].apply(np.float64)
df.dtypes
```

# 4.Duplicates

We can identify which observations are duplicates
through the duplicated() function and sum() to know how many duplicate records there are. 

```python
nbadf = pd.read_csv('http://bit.ly/NBABasketballDataset')
nbadf.shape
```

```python
nbadf2 = nbadf[nbadf.duplicated()]
nbadf2.head()
```

```python
sum(nbadf.duplicated())
```

```python
df_duplicates = nbadf.drop_duplicates()
df_duplicates
```

# 6.Outliers

```python
outliers_df = pd.read_csv('http://bit.ly/CountryDataset1')
outliers_df.head()
```

```python
outliers_df.shape
```

```python
Q1 = outliers_df.quantile(0.25)
Q3 = outliers_df.quantile(0.75)
IQR = Q1 - Q1
IQR
```

```python
outliers_df_iqr = outliers_df[((outliers_df < (Q1 - 1.5 * IQR)) | (outliers_df > (Q3 + 1.5 * IQR))).any(axis=1)]
outliers_df_iqr.shape
```
